# GCD Time Analysis

```python
m = 625
n = 6215
```

```python
x = [m]
y = [n]

print(m,n)

while (m%n!=0):
    temp = m
    m = n%m
    n = temp
    
    x.append(m)
    y.append(n)
    
    print(m,n)

import numpy as np 
x = np.array(x)
y = np.array(y)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.scatter(x, y)

plt.xlabel('n', fontsize = 20)
plt.ylabel('m', fontsize = 20)
plt.show()
```

# Log Testing 

```python
num = 10000

val = []

for i in range(1, num):
    # print("Stage ", i)
    
    for j in range(i, num):
        val.append( (math.log(j,2))  )   
        
val.sort()
```

# Steps Testing 

```python
def gcd_steps(n,m):
    count = 0
    while (m%n!=0):
        temp = m
        m = n%m
        n = temp

        count += 1

    return count
```

```python
num = 10000

steps = []

for i in range(1, num):    
    for j in range(i, num):
        steps.append(gcd_steps(j, i))
        
steps.sort()
```

```python
y = np.array(steps)
# x = np.array([i for i in range( (num * (num-1)) //2 ) ])

plt.plot(y,x)
plt.xlabel("No of iterations - steps")
plt.ylabel("Tested value ranges")
plt.show() 
```
